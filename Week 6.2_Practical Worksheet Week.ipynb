```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
#this will make your plot outputs appear and be stored within the notebook
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic['Age'].hist()
```

```python
titanic['Age'].hist(bins=20)
```

```python
import seaborn as sns
sns.set_style('darkgrid')
titanic['Age'].hist(bins=20)
```

```python
titanic.plot.line(x='Age',y='Fare',figsize=(8,6))
```

```python
titanic.plot.scatter(x='Age',y='Fare',figsize=(8,6))
```

```python
titanic.plot.box(figsize=(10,8))
```

```python
titanic.plot.hexbin(x='Age',y='Fare',gridsize=30,figsize=(8,6))
```

```python
titanic['Age'].plot.kde()
```

```python
titanic['Pclass'].value_counts().sort_index().plot.bar()
```

```python
titanic['Pclass'].value_counts().sort_index().plot.barh()
```

```python
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
ax = sns.countplot(titanic['Survived'],hue=titanic['Sex'],palette='Greys')
plt.xticks(rotation = 30, ha = 'right')
```
